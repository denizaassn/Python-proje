```python
class Hesap:
    def __init__(self,hesap_no,ad_soyad,sifre,bakiye):
        self.hesap_no = hesap_no
        self.ad_soyad = ad_soyad
        self._sifre = sifre
        self._bakiye = bakiye
        self._log = []

    def dogrula(self,girilen_sifre):
        return self._sifre == girilen_sifre

    
    def para_yatir(self,miktar):
        if miktar > 0:
            self._bakiye += miktar
            self._log.append(f"{miktar} TL yatırıldı.")
            print(f"{miktar} TL yatırıldı. Yeni bakiye: {self._bakiye} TL")
        else:
            print("Geçersiz miktar.")


    def para_cek(self,miktar):
        if 0 < miktar <= self._bakiye:
            self._bakiye -= miktar
            self._log.append(f"{miktar} TL çekildi.")
            print(f"{miktar} TL çekildi. Kalan bakiye: {self._bakiye} TL")
        else:
            print("Yetersiz bakiye ya da geçersiz miktar.")

    def bakiye_goster(self):
        print(f"{self.ad_soyad} adlı kullanıcının bakiyesi: {self._bakiye} TL")



    def sifre_degistir(self,eski_sifre,yeni_sifre):
        if self._sifre == eski_sifre:
            self._sifre = yeni_sifre
            print("Şifre başarıyla değiştirildi.")
            self._log.append("Şifre değiştirildi.")
        else:
            print("Eski şifre yanlış.")

    
    def log_goster(self):
        print("\n Hesap Hareketleri:")
        for hareket in self._log:
            print("-",hareket)
        print()

# Menü Sistemi 

def atm_menusu(hesap):
    while True:
        print("\n İşlem Menüsü:")
        print("1. Bakiye Görüntüle")
        print("2. Para Yatır")
        print("3. Para Çek")
        print("4. Hesap Hareketlerini Göster")
        print("5. Şifre Değiştir")
        print("6. Çıkış")
        
        secim = input("Seçiminizi girin (1-6): ")

        if secim == "1":
            hesap.bakiye_goster()
            
        elif secim == "2":
            miktar = float(input("Yatırmak istediğiniz miktar:"))
            hesap.para_yatir(miktar)
            
        elif secim == "3":
            miktar = float(input("Çekmek istediğiniz miktar:"))
            hesap.para_cek(miktar)
            
        elif secim == "4":
            hesap.log_goster()

        elif secim == "5":
            eski = input("Eski şifre:")
            yeni = input("Yeni şifre:")
            hesap.sifre_degistir(eski,yeni)
            
        elif secim == "6":
            print("Güle güle!")
            break
            
        else:
            print("Geçersiz seçim.")


# Giriş Ekranı

def giris_ekrani(hesaplar):
    print("ATM Giriş Sİstemi")
    hesap_no = input("Hesap numaranızı girin:")
    if hesap_no in hesaplar:
        hesap = hesaplar[hesap_no]
        for _ in range(3):
            sifre = input("Şifrenizi girin:")
            if hesap.dogrula(sifre):
                print(f"Hoş geldiniz, {hesap.ad_soyad}!")
                atm_menusu(hesap)
                return
            else:
                print("Hatalı şifre.")
        print("3 kez hatalı giriş yapıldı. Sistemden çıkılıyor.")
    else:
        print("Hesap bulunamadı.")




# Örnek Hesap

hesaplar = {
    "123": Hesap("123","Deniz Asan","1234",5000),
    "456": Hesap("456","Ece Kavak","4567",4000)
}



while True:
    giris_ekrani(hesaplar)
    tekrar = input("Başka bir kullanıcı giriş yapacak mı ? (evet/hayır):").lower()
    if tekrar != "evet":
        print("ATM sistemi kapatılıyor...")
        break
    
```
